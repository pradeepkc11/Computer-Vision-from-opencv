# Rotation of Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image=cv2.imread("index.png")
rows,cols=image.shape[:2]
m=cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst=cv2.warpAffine(image,m,(cols,rows))
plt.imshow(dst)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```
